```python
import googlemaps

gmaps = googlemaps.Client(key='///')

def find_address(shelter_name, borough):
    if pd.isnull(shelter_name) or pd.isnull(borough):
        return None
    location = f"{shelter_name}, {borough}, NYC"
    geocode_result = gmaps.geocode(location)
    if geocode_result:
        return geocode_result[0]['geometry']['location']
    else:
        return None

# Apply the function to the 'ShelterName' column
df['location'] = df.apply(lambda row: find_address(row['ShelterName'], row['Borough']), axis=1)

# Split location into latitude and longitude
df['latitude'] = df['location'].apply(lambda loc: loc['lat'] if loc else None)
df['longitude'] = df['location'].apply(lambda loc: loc['lng'] if loc else None)
```
